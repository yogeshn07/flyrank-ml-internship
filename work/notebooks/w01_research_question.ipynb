## 2. The question: decision, action, cost of a wrong call

## Research question

**Decision:** Which content pages should be reviewed first for potential refresh?

**Action:** A content reviewer uses the ranked list to decide which pages to inspect before updating or expanding content.

**Cost of a wrong recommendation:** A false positive wastes reviewer time on pages that do not need attention, while a false negative may delay improvements for pages that are actually declining. The objective is therefore to prioritize review efficiently rather than guarantee that refreshing a page will improve its performance.

Machine learning can help because multiple observable signals—such as impressions, CTR, average position, content age, and update history—may collectively identify stronger review candidates than simple hand-written rules.

## 3. Quick look at the data (2-3 real numbers)

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

print(f"Total pages: {len(df):,}")

declining_rate = (df["trend_direction"] == "down").mean()

print(f"Declining pages: {declining_rate:.3f}")

corr = df["search_volume"].corr(df["impressions_90d"])

print(f"Correlation between search volume and impressions: {corr:.3f}")

visible = df[df["impressions_90d"] >= 100]

median_ctr = visible.groupby("position_tier")["ctr"].median()

print("\nMedian CTR by position tier:")

print(median_ctr.round(3))
```
